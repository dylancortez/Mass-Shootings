# Effects of COVID 19 on Mass Shootings in the United States

## Introduction
This report represents the results of a data analysis regarding mass shootings in the United States from 2017 to 2022. The purpose of this analysis is to determine whether the United State's lockdowns and COVID restrictions had a tangible impact on the occurrence of mass shootings in the country.

## Data Collection and Cleaning
The data used in this analysis was collected from <a>gunviolencearchive.org</a>, an independent research and data collection organization created to provide comprehensive data for gun violence. The data was cleaned and processed using Python's Pandas library, which involved removing any duplicate or missing values, and transforming the data into a usable format for analysis. For this report, a mass shooting is defined as <i>"any gun violence incident with a minimum of four victims shot, either injured or killed, not including any shooter who may have also been killed or injured in the incident."</i>

## Methodology
To interpret the data, we used a series of line charts and scatter plots to track trends in mass shootings over time. We performed the analysis using Python's NumPy library, and graphed the information using Python's Matplotlib library.

## Results
Based on the data provided, we cannot determine that COVID restrictions (which started in the US on March 12, 2020) had any impact whatsoever on the frequency of mass shootings across the nation. In fact, 2020 saw an exponential increase in the number of mass shootings.

## Limitations
The data gone over in this report only shows the frequency of mass shootings across the United States in total, and does not delve into the cities and states where COVID restrictions may have been harsher than other parts of the country. This is only a high-level overview of the statistics given.
